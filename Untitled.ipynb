```scijava
//load ImageJ
%classpath config resolver imagej.public https://maven.imagej.net/content/groups/public
%classpath add mvn net.imagej imagej 2.0.0-rc-68

//create ImageJ object
ij = new net.imagej.ImageJ()
```

```scijava
chNuclear_n = 0
ch1_n = 1
ch2_n = 2

voxSize = 0.16 * 0.16 * 0.24; // um^3

nuclei_dog = true
nuclei_mip = false      // Coloc: True, Single: False
nuclei_sigma = [2,2,1]  // Coloc: [10 10 3], Single: [2 2 1]
nuclei_bkg_sigma = [69,69,12]
c1_tophat = 10
c2_tophat = 10
c1_sigma = [1.5,1.5,1]
c2_sigma = [1.5,1.5,1]
filterBySize = false
singleCell = true
```

```scijava
//image = ij.io().open("example_data/test1.tif")
image = ij.scifio().datasetIO().open("example_data/test1.tif")
w = image.dimension(0)
h = image.dimension(1)
ch = image.dimension(2)
d = image.dimension(3)

[w,h,ch,d]
```

```scijava
import net.imglib2.FinalInterval

// im[:,:,z]
getZSliceXYZ = { im, z -> ij.op().run("crop" , im, FinalInterval.createMinSize( 
                                         0, 0, z as long,
                                         im.dimension(0), im.dimension(1), 1), true) }
```

```scijava
getMid = { im -> getZSliceXYZ(im,im.dimension(2)/2) }
```

```scijava
import net.imglib2.FinalInterval

(nuc, ch1, ch2) = [chNuclear_n, ch1_n, ch2_n].stream().map { i -> ij.op().run(
    "crop", 
    image, 
    FinalInterval.createMinSize(0,0,i,0,w,h,1,d),
    true) } .collect()
```

```scijava
[nuc,ch1,ch2].stream().map { a -> getMid(a) } .collect()
```

```scijava
ops = ij.op()

ch1_mid = getMid(ch1)

ch1_mid.numDimensions()
```

```scijava
out = ops.run("create.img", ch1_mid, new net.imglib2.type.numeric.real.DoubleType())

vars = [
    "out" : ops.run("create.img", ch1_mid, new net.imglib2.type.numeric.real.DoubleType()),
    "dxx" : ops.run("create.img", ch1_mid, new net.imglib2.type.numeric.real.DoubleType()),
    "dyy" : ops.run("create.img", ch1_mid, new net.imglib2.type.numeric.real.DoubleType()),
    "ch1_mid" : ch1_mid,
    "dx2" : [2,0] as int[],
    "dy2" : [0,2] as int[],
    "sigma" : 3
]
out = ops.eval(" invert(out,derivativeGauss(dxx, ch1_mid, dx2, sigma) + derivativeGauss(dyy, ch1_mid, dy2, sigma))", vars)

ij.notebook().display(out)
```

```scijava
ops.run("filter.derivativeGauss", out, ch1_mid, [0,2] as int[], 3)
ops.run("filter.derivativeGauss", out, ch1_mid, [0,2] as int[], 3)

ij.notebook().display(out)
```
